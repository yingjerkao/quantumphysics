#The Bohr-Sommerfeld quantization rule


## Bohr's model of the atoms
Consider an atom consisting of a nucleus of charge $+Ze$ and mass $M$, and a single electron of charge $-e$ and mass $m$.

We assume that the electron revolves in a circular orbit about the nucleus. We assume that the nucleus remains fixed in space. The condition of mechanical stability of the electron is

\begin{equation}
\frac{1}{4\pi\epsilon_0}\frac{Ze^2}{r^2}=m\frac{v^2}{r}
\end{equation}


## Sommerfeld quantization rule

\begin{align}
\oint L d\theta &=n_\theta h\\
\oint p_r d\theta &=n_r h
\end{align}

The first condition yields the same restriction on the orbital angular momentum

\begin{equation}
L=n_\theta \hbar,\quad\quad n_\theta=1,2,3,\ldots
\end{equation}

The second condition leads to the following relation between $L$ and $a/b$, the ratio of the semimajor axis $a$ to the semiminor axis $b$ of the ellipse,

\begin{equation}
L(a/b-1)=n_r \hbar,\quad\quad n_r=0, 1, 2, 3,\ldots
\end{equation}

For an elliptic orbit in a Coulomb potential,
\begin{align}
a&=-\frac{1}{4\pi\epsilon_0}\frac{Ze^2}{2E}\\
b&=\frac{L}{\sqrt{-2\mu E}}\\
1-\epsilon^2&=-\frac{2E L^2}{\mu}\left(\frac{4\pi\epsilon_0}{Ze^2}\right)^2, 
\end{align}
where $b/a=\sqrt{1-\epsilon^2}$.

Applying the quantization rules, 
\begin{align}
a&=\frac{4\pi\epsilon_0 n^2\hbar^2}{\mu Z e^2}\\
b&=a\frac{n_\theta}{n}\\
E&=-\left(\frac{1}{4\pi\epsilon_0}\right)^2 \frac{\mu Z^2e^4}{2 n^2\hbar^2},
\end{align}
where $n=n_r+n_\theta=1,2,3,4,\ldots$ is called the *principal quantum number*, and $n_\theta=1,2,3,\ldots, n$ is called the *azimuthal quantum number*.

```python
%pylab inline
from __future__ import division
from matplotlib.patches import Ellipse
from IPython.html.widgets import interactive
from IPython.html import widgets

def Sommerfeld_orbits(n=1):
    fig = figure(figsize=(16,16))
    ax = fig.add_subplot(111, aspect='equal')
    a=n**2
    ells=[]
    for n_theta in range(1,n+1):
        b=a*n_theta/n
        e=sqrt(1-(b/a)**2)
        xcenter,ycenter=a*e, 0
        ells.append(Ellipse(xy=(xcenter,ycenter),width=2*a, height=2*b))
        
    for e in ells:
        ax.add_artist(e)
        e.set_clip_box(ax.bbox)
        e.set_alpha(0.5)
        e.set_facecolor('none')
    ax.annotate('$n=$'+str(n), xy=(.2, .2),  xycoords='axes fraction',fontsize=18)
    ax.set_xlim(-50,50)
    ax.set_ylim(-50,50)


    
        
w=interactive(Sommerfeld_orbits,n=(1,10,1))
display(w)
```
