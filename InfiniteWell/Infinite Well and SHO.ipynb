#Infinite Well

\begin{align}
H\psi(x)&=E\psi(x)\\
H&=\frac{p^2}{2m}+ V(x),\quad V(x)=\left\{
\begin{split}
&0, \quad 0<x<a;\\
&\infty, \quad \mbox{otherwise}.
\end{split}
\right.
\end{align}

```python
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline
from IPython.html.widgets import interact, interactive
from IPython.html import widgets
from IPython.display import display, Math, Latex
from __future__ import division
from sympy import *
import sympy.plotting as splt

init_printing(use_latex='mathjax')
```

```python
var('x,a,y', real=True, positive=True)

def phi(n):
    return sqrt(2/a)*sin(n*pi*x/a)

def phi2(n,a):
    return sqrt(2/a)*sin(n*pi*x/a)

def eigen(n):
    p = splt.plot_parametric((0, y),(y,0, 2), line_color='black',show=False)
    p.extend(splt.plot_parametric((5, y),(y,0, 2), line_color='black',show=False))
    p.extend(splt.plot_parametric((2.5, y),(y,0, 2), line_color='r',show=False))
    display( Math("\phi(x)="),phi(n))
    q=plot(phi2(n,5),(x,0,5),show=False)
    p.extend(q)
    p.show()
interact(eigen,n=(1,10));

```

###Eigenfunctions are orthonormal

\begin{equation}
\int_0^a \psi_m(x) \psi_n(x) dx =\delta_{mn}
\end{equation}

###Example 1:
$m=1$, $n=2$

```python
display(phi(1))
display(phi(2))
eg1=Integral(phi(1)*phi(2),(x,0,a))
display(Eq(eg1,eg1.doit()))
```

###Example 2:
$m=n=2$


```python

eg2=Integral(phi(2)*phi(2),(x,0,a))
display(Eq(eg2,eg2.doit()))
```

```python
colors=['r','g','b','y','c','m','k']
p = splt.plot_parametric((0, y),(y,0, 1), line_color='k',show=False)
p.extend(splt.plot_parametric((5, y),(y,0, 1), line_color='k',show=False))
p.save('tmp.png')
fig=p._backend.fig
ax=p._backend.ax
xt=np.linspace(0,5,100)


fig.set_size_inches(6,8)
for n in range(1,5):
    E=n**2/(2*5*5)
    f=lambdify(x,phi2(n,5),'numpy')
    ax.plot(xt,f(xt)+E,color=colors[Mod(n-1,7)])

    ax.set_xlabel("$x$",fontsize=18)
    ax.set_ylabel("$\psi(x), E$",fontsize=18)
    ax.set_title('Infinite Well',fontsize=18) 
    ax.annotate('$\Psi_{%s}$, $E={%s}^2E_1$' %(n,n),(5.2,E), xycoords='data',size=16)
    ax.fill(xt,f(xt)+E,'lightgrey',alpha=0.5)
```

###Simple Harmonic Oscillator

```python
def Harm_Osc(n):
    return sqrt(1/(2**n*factorial(n)))/(pi)**(0.25)*exp(-x**2/2)*hermite(n,x)

xpts=np.linspace(-3.2,3.2,1000)
fig1=plt.figure(figsize=(12,8))
ax1=fig1.add_subplot(111)
ax1.set_xlabel("$x$",fontsize=18)
ax1.set_ylabel("$\psi(x), E$",fontsize=18)
ax1.set_title('Simple Harmonic Oscillator',fontsize=18) 
ax1.plot(xpts, xpts**2/2 ,color='k')
ax1.annotate('$x^2/2$',(3.2,5), xycoords='data',fontsize=13)
xpts=np.linspace(-5,5,1000)
for n in range(5):
    f=lambdify(x,Harm_Osc(n)+n+0.5,'numpy')
    ax1.plot(xpts,f(xpts),colors[n])
    ax1.annotate('$\Psi_{%s}$' %n, (5.2, n+0.5),xycoords='data',fontsize=15)
    , $E=(%s+1/2)\hbar\omega_0$
    ax1.annotate('$E=(%s+1/2)\hbar\omega_0$' %n, (-5, n+0.9),xycoords='data',fontsize=15)
    , 
    ax1.fill(xpts,f(xpts),'grey',alpha=0.2)
plt.show()
```
