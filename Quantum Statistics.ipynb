```python
k0=8.6173324E-5
def Boltzmann(e,mu,T):
    return exp(-(e-mu)/k0/T)
def Bose(e,mu,T):
    return 1/(exp((e-mu)/k0/T)-1)
def Fermi(e,mu,T):
    return 1/(exp((e-mu)/k0/T)+1)
```

```python

E=arange(0.01,3,0.01)
fig1=plot(E,Boltzmann(E,2.84*k0*1000,1000))+plot(E,Boltzmann(E,0.42*k0*5000,5000))+plot(E,Boltzmann(E,-0.62*k0*10000,10000))
ylim([0,2])
xlabel('$\epsilon$',fontsize='xx-large')
ylabel('n($\epsilon$)',fontsize='xx-large')
show(fig1)
```

```python
fig2=plot(E,Bose(E,0,1000))+plot(E,Bose(E,0,5000))+plot(E,Bose(E,0,10000))
ylim([0,2])
xlabel('$\epsilon$',fontsize='xx-large')
ylabel('n($\epsilon$)',fontsize='xx-large')
show(fig2)
```

```python
fig3=plot(E,Fermi(E,2,1000))+plot(E,Fermi(E,2,5000))+plot(E,Fermi(E,2,10000))+plot(E,Fermi(E,2,0.001))
ylim([0,1.5])
xlabel('$\epsilon$',fontsize='xx-large')
ylabel('n($\epsilon$)',fontsize='xx-large')
show(fig3)
```
